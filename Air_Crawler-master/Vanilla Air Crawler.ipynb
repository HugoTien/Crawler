```python
!pip install splinter
import splinter
import time
from splinter import Browser
import datetime
from datetime import datetime, timedelta
import os
import random
```

```python
browser = splinter.Browser(driver_name='chrome',executable_path='D:/chorme/chromedriver ')
browser.driver.set_window_size(1680, 1050)

for i in range(1,180):
    try:
        browser.cookies.delete()
        browser.visit('https://www.vanilla-air.com/tw/')
        browser.click_link_by_id("backAndForth")

        time.sleep(1)

        button = browser.find_by_css('.minict_wrapper.active')[0].click()
        time.sleep(1)
        browser.find_by_text('台北（桃園）').first.click()
        time.sleep(1)
        time.sleep(1)
        browser.find_by_text('大阪（關西）')[1].click()
        time.sleep(3)

        date = str(datetime.today() + timedelta(days=i))
        date = date.split(" ")[0].replace("-","/")

        browser.fill('travelDate1[date]', date)
        time.sleep(2)
        browser.fill('travelDate2[date]', date)
        browser.find_by_xpath('//*[@id="adultPeopleBox"]').click() # 隨便點個東西
        time.sleep(2)
        browser.find_by_id("edit-submit-ticket").first.click()
        time.sleep(5)


        filename = date.replace("/","_")

        print(str(date)+" Done")
        with open('C:/Users/Java/Desktop/Project/Air_Crawler/香草html/{}.html'.format(filename),'w',encoding="utf-8") as f:
            f.write(browser.html)
        time.sleep(random.randrange(5,10))
    except:
        pass

```
