```python
# real
import requests
from bs4 import BeautifulSoup
from collections import Counter
import re
import threading
import time


joblist = ['程式設計師']



linklist = []


def getlinklist():
    for job in joblist:   
        url = "https://www.518.com.tw/job-index-P-{}.html?i=1&am=1&ad={}&xq=1".format(1,'程式設計師')
        r1 = requests.get(url)
        soup = BeautifulSoup(r1.text,'lxml')
        # 初步設定
        getpage = soup.find_all('span','pagecountnum')
        totalpage = int(getpage[0].text.split("/ ")[1].split("頁")[0])
        # 確定最大 page
        for page in range(1,totalpage+1): # 6為第6頁，實際請填上totalpage
            url = "https://www.518.com.tw/job-index-P-{}.html?i=1&am=1&ad=%{}&xq=1".format(page,job)
            r1 = requests.get(url)
            soup = BeautifulSoup(r1.text,'lxml')
            link = soup.select('.title > a') 
            for joblink in link:
                linklist.append(joblink['href'])

    return linklist

getlinklist()


skills = []



for joblink in linklist:
    joburl = joblink
    r2 = requests.get(joblink)
    soup = BeautifulSoup(r2.text,'lxml')
    skill = soup.find_all('div','job-detail-box')
    for skillitem in skill:
        try:
            skills.append("".join(skillitem.text.replace("\n","").split("擅長工具：")[1].split("工作技能：")[0]))
        except IndexError:
            pass
skills

```

```python
sklist = []
for subsk in skills:
    sklist.append(subsk.strip().replace("/",",").replace("、",",").split(","))
sklist
```

```python
skcount = Counter()
for sk in sklist:
    for subsk in sk:
        if subsk in skcount:
            skcount[subsk]+=1
        else:
            skcount[subsk]=1


skcount.most_common(200)
```

```python
import json
json.dump(skcount,open("skcount.json","w"),ensure_ascii=False)
```
